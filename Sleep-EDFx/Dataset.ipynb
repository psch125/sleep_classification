```python
from PIL import Image
import os, glob, numpy as np
from sklearn.model_selection import train_test_split
import os, glob, numpy as np

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
%matplotlib inline 
caltech_dir = "C:/Users/PC/표/cnn_eeg/train_data2"
# categories = ["1_emg", "2_emg", "3_emg","R_emg"]
categories =["1_eeg_fft+eog_fft","2_eeg_fft+eog_fft", "3_eeg_fft+eog_fft","R_eeg_fft+eog_fft"]
nb_classes = len(categories)

image_w = 216
image_h = 144

pixels = image_h * image_w * 3

X = []
y = []


for idx, cat in enumerate(categories):
    
    #one-hot 돌리기.
    label = [0 for i in range(nb_classes)]
    label[idx] = 1

    image_dir = caltech_dir + "/" + cat
    print(image_dir)
    files = glob.glob(image_dir+"/*.png")
    print(cat, " 파일 길이 : ", len(files))
    for i, f in enumerate(files):
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)

        X.append(data)
        y.append(label)

        if i % 70 == 0:
            print(cat, " : ", f)

X = np.array(X)
y = np.array(y)

#1 0 0 0 이면 airplanes
#0 1 0 0 이면 buddha 이런식

# X_train, X_test, y_train, y_test = train_test_split(X,y , shuffle = False, test_size = 0.1)
X_train, X_test, y_train, y_test = train_test_split(X, y)
xy = (X_train, X_test, y_train, y_test)

np.save("C:/Users/PC/표/cnn_eeg/sklearn_dataset2/eeg_fft+eog_fft.npy", xy)




```
